```python
import matplotlib.pyplot as plt
import numpy as np  

# Graph 1
plt.subplot(3, 3, 1)
plt.title("Title 1")
x= np.array([1, 8])
y= np.array([3, 10])
plt.plot(x,y,"o", c="blue")

plt.xlabel("Y label")
plt.ylabel("X label")



# Graph 2
plt.subplot(3, 3, 2)
plt.plot([1, 3, 5, 7, 9], [2, 6, 2, 6, 2], "*", mfc="orange", ls="-",lw=2.5)
plt.plot([1, 3, 5, 7, 9], [6, 3, 5, 3, 6], ls="dashed",lw=2.5)

# Graph 3
plt.subplot(3, 3, 3)
plt.plot([1, 3, 5, 7], [3, 6, 3, 6], marker="o", mfc="yellow", mec='black', c="blue", ls="-")

# Graph 4
plt.subplot(3, 3, 4)
plt.title("Table 2")
plt.plot([1, 3, 5, 7], [1, 3, 5, 7])
plt.grid()



#graph 5
plt.subplot(3, 3, 5)
x = np.random.randint(25, size=(100))
y = np.random.randint(20, size=(100))
colors = np.random.randint(30, size=(100))
sizes = 10 * np.random.randint(22, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5)



#graph 6

plt.subplot(3, 3, 6)
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y,width=0.1)
plt.suptitle("GRAPH")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 20, 20])

plt.pie(y)
plt.show() 
```

```python
y = np.array([35, 25, 20, 20])
myexplode = [0.2, 0, 0, 0]
mycolors = ["blue", "hotpink", "orange", "#4CAF50"]
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y,startangle = 90, explode = myexplode, shadow = True,colors=mycolors,labels = mylabels)
plt.legend()
plt.show() 
```
