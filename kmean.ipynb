```python
import numpy as np
from sklearn import datasets
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data     # Features
y = iris.target   # Ground truth labels


k = 3

# Initialize the KMeans model

kmeans = KMeans(n_clusters=k,n_init='auto')
kmeans.fit(X)

# Get the cluster centers and labels

cluster_centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Visualize the data points and cluster centers

plt.figure(figsize=(12, 6))

# Scatter plot for the first two features (sepal length and sepal width)

plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], marker='x', s=200, color='red')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Features')

# Scatter plot for the last two features (petal length and petal width)

plt.subplot(1, 2, 2)
plt.scatter(X[:, 2], X[:, 3], c=labels, cmap='viridis')
plt.scatter(cluster_centers[:, 2], cluster_centers[:, 3], marker='x', s=200, color='red')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Petal Features')
plt.show()
```
