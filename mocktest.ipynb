```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score

# Load the Iris dataset
iris = load_iris()
X = iris.data       
y = iris.target     

# Split dataset into training and testing data (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create a k-NN classifier with k=3
knn = KNeighborsClassifier(n_neighbors=3)

# Train the classifier
knn.fit(X_train, y_train)

# Make predictions on the test set
y_pred = knn.predict(X_test)

# Evaluate the classifier
conf_matrix = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)

# Output results
print("Confusion Matrix:")
print(conf_matrix)
print("\nAccuracy of k-NN classifier: {:.2f}%".format(accuracy * 100))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LinearRegression
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, mean_squared_error, r2_score

# 1. k-NN Classification using Iris dataset
iris = datasets.load_iris()
X_knn, y_knn = iris.data, iris.target
X_train_knn, X_test_knn, y_train_knn, y_test_knn = train_test_split(X_knn, y_knn, test_size=0.3, random_state=1)
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_knn, y_train_knn)
y_pred_knn = knn.predict(X_test_knn)
cm_knn = confusion_matrix(y_test_knn, y_pred_knn)
accuracy_knn = accuracy_score(y_test_knn, y_pred_knn)
print("1. k-NN Accuracy:", accuracy_knn)
print("Confusion Matrix:\n", cm_knn)

# 2. Plotting Electricity Power Consumption
years = [1972, 1982, 1992, 2002, 2012]
e_india = [100.6, 158.61, 305.54, 394.96, 724.79]
e_bangladesh = [10.5, 25.21, 58.65, 119.27, 274.87]

plt.figure()
plt.plot(years, e_india, color='blue', marker='o', label='India')
plt.plot(years, e_bangladesh, color='green', marker='s', label='Bangladesh')
plt.xlabel('Year')
plt.ylabel('Electricity Power Consumption (TWh)')
plt.title('Electricity Power Consumption: India vs Bangladesh')
plt.legend()
plt.grid(True)
plt.show()

# 3. Naive Bayes Classification
X_nb, y_nb = iris.data, iris.target
X_train_nb, X_test_nb, y_train_nb, y_test_nb = train_test_split(X_nb, y_nb, test_size=0.3, random_state=1)
gnb = GaussianNB()
gnb.fit(X_train_nb, y_train_nb)
y_pred_nb = gnb.predict(X_test_nb)
cm_nb = confusion_matrix(y_test_nb, y_pred_nb)
accuracy_nb = accuracy_score(y_test_nb, y_pred_nb)
print("\n3. Naive Bayes Accuracy:", accuracy_nb)
print("Confusion Matrix:\n", cm_nb)

# 4. Dictionary to DataFrame
data_dict = {
    'Name': ['Jai', 'Navin', 'Mohan', 'Anuj'],
    'Age': [27, 24, 28, 32],
    'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Chennai'],
    'Qualification': ['Msc', 'MA', 'MCA', 'PhD']
}
df = pd.DataFrame(data_dict)
df['Name'] = df['Name'] + ' - ' + df['Address']  # Overwrite Name column
print("\n4. Modified DataFrame:")
print(df)

# 5. Linear Regression using Diabetes dataset
diabetes = datasets.load_diabetes()
X_lr, y_lr = diabetes.data, diabetes.target
X_train_lr, X_test_lr, y_train_lr, y_test_lr = train_test_split(X_lr, y_lr, test_size=0.3, random_state=42)
lr = LinearRegression()
lr.fit(X_train_lr, y_train_lr)
y_pred_lr = lr.predict(X_test_lr)
mse_lr = mean_squared_error(y_test_lr, y_pred_lr)
r2_lr = r2_score(y_test_lr, y_pred_lr)
print("\n5. Linear Regression MSE:", mse_lr)
print("Linear Regression R²:", r2_lr)

# 6. Pie chart of car data
cars = ['AUDI', 'BMW', 'FORD', 'TESLA', 'JAGUAR', 'MERCEDES']
data = [23, 17, 35, 29, 12, 41]
plt.figure()
plt.pie(data, labels=cars, autopct='%1.1f%%', startangle=140)
plt.title('Car Distribution')
plt.axis('equal')
plt.show()

# 7. Decision Tree with sample prediction
dt = DecisionTreeClassifier(random_state=0)
dt.fit(X_train_nb, y_train_nb)
y_pred_dt = dt.predict(X_test_nb)
accuracy_dt = accuracy_score(y_test_nb, y_pred_dt)
sample_instance = [[5.1, 3.5, 1.4, 0.2]]
sample_prediction = dt.predict(sample_instance)[0]
print("\n7. Decision Tree Accuracy:", accuracy_dt)
print("Prediction for Sample [5.1, 3.5, 1.4, 0.2]:", iris.target_names[sample_prediction])

# 8. Array operations
arr = np.array([1, 2, 3, 4, 5])
print("\n8. Array operations:")
print("Original Array:", arr)
print("Add 1:", arr + 1)
print("Subtract 2:", arr - 2)
print("Sum of elements:", arr.sum())
```
