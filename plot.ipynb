```python

x = np.array([5,2, 10,3])
y = np.array([3, 8, 6, 10])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)


plt.grid(axis='y',c='green')
plt.plot(x, linestyle = 'dashed')
plt.plot(y, linewidth = '1')
plt.plot(y,marker='o',mec='r',ms='8',lw='2.5',mfc='b')
plt.show()
```

```python
x = np.array([5,2, 10,3])
y = np.array([3, 8, 6, 10])
plt.ylabel("y axis")

plt.subplot(2, 2, 1)
plt.plot(y)
plt.plot(x)
plt.title("1")


plt.subplot(2, 2, 2)
plt.plot(x)
plt.plot(y)
plt.title("2")
```
